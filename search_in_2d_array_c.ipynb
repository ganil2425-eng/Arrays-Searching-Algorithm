```python
%%writefile matrix_search.c
#include <stdio.h>

int main()
{
    int n1, n2;
    printf("Enter the 2D array first row size: ");
    scanf("%d", &n1);

    printf("Enter the 2D array second row size: ");
    scanf("%d", &n2);

    int ar[n1][n2];

    // Input elements
    for(int i = 0; i < n1; i++)
    {
        for(int j = 0; j < n2; j++)
        {
            printf("Enter element [%d][%d]: ", i, j);
            scanf("%d", &ar[i][j]);
        }
    }

    int search;
    int flag = 0;

    printf("Enter the number to search: ");
    scanf("%d", &search);

    // Search element
    for(int i = 0; i < n1; i++)
    {
        for(int j = 0; j < n2; j++)
        {
            if(ar[i][j] == search)
            {
                flag = 1;
                break; // exit inner loop
            }
        }
        if(flag == 1) break; // exit outer loop
    }

    if(flag == 1)
    {
        printf("Element found!\n");
    }
    else
    {
        printf("Element not found.\n");
    }

    return 0;
}
```
